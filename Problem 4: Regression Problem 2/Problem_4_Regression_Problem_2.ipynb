```python
import pandas as pd
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Load the training and testing data
train_data = pd.read_csv('p2_train.csv')
test_data = pd.read_csv('p2_test.csv')
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
# Separate features and target variable for training and testing
X_train = train_data.iloc[:, :-1]  # Features in training data
y_train = train_data.iloc[:, -1]   # Target variable in training data

X_test = test_data.iloc[:, :-1]    # Features in testing data
y_test = test_data.iloc[:, -1]     # Target variable in testing data
```

```python
# Initialize SVR model
svr_model = SVR(kernel='rbf')
```

```python
# Train SVR model on the training data
svr_model.fit(X_train, y_train)
```

```python
# Initialize Linear Regression model
linear_model = LinearRegression()
```

```python
# Train Linear Regression model on the training data
linear_model.fit(X_train, y_train)
```

```python
# Predict the target variable on the test set for SVR and Linear Regression
svr_predictions = svr_model.predict(X_test)
linear_predictions = linear_model.predict(X_test)
```

```python
# Calculate Mean Squared Error and Mean Absolute Error for SVR
svr_mse = mean_squared_error(y_test, svr_predictions)
svr_mae = mean_absolute_error(y_test, svr_predictions)
```

```python
# Calculate Mean Squared Error and Mean Absolute Error for Linear Regression
linear_mse = mean_squared_error(y_test, linear_predictions)
linear_mae = mean_absolute_error(y_test, linear_predictions)
```

```python
# Print the metrics for SVR and Linear Regression
print("SVR Metrics:")
print("Mean Squared Error:", svr_mse)
print("Mean Absolute Error:", svr_mae)

```

```python
print("\nLinear Regression Metrics:")
print("Mean Squared Error:", linear_mse)
print("Mean Absolute Error:", linear_mae)
```
