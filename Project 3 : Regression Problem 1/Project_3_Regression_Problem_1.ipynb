```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Load the training and testing data
test_data = pd.read_csv('/content/test.csv')
train_data = pd.read_csv('/content/train.csv')
```

```python
# Separate features and target variable for training and testing
X_train = train_data.iloc[:, :-1]  # Features in training data
y_train = train_data.iloc[:, -1]   # Target variable in training data

X_test = test_data.iloc[:, :-1]    # Features in testing data
y_test = test_data.iloc[:, -1]     # Target variable in testing data
```

```python
# Initialize the linear regression model
model = LinearRegression()
```

```python
# Train the model on the traing data
model.fit(X_train, y_train)
```

```python
# Predict the target variable on the test set
predictions = model.predict(X_test)
```

```python
# Calculate Mean Squared Error and Mean Absolute Error
mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)
```

```python
# Print the metrics
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
```
